#### Добавляем столбец с названиями когорт

#### Добавляем столбец с количеством реакций

```python
# функция, возрвщающая количество реакций
def sum_reactions(name):
    try:
        a=name
        b=re.findall(r"'count': (\d+)", a) # возвращаем совпадения шаблона
        l=[]
        for ch in b:
            if ch.isdigit(): # проверяем числа
                l.append(int(ch)) # добавляем значение в виде int
        return sum(l) # возвращаем сумму значений
    except:
        return 0

df['cnt_reactions'] = df['reactions'].apply(sum_reactions)

pd.set_option('display.max_colwidth', None) # сделаем так, чтобы значение столбца было видно полностью
df.loc[7,['reactions','cnt_reactions']]
```

#### Добавляем столбец с названиями каналов сообщений

Выделим основные каналы: *info, teamwork, exerciser, projects*; остальные определим как *other*.

```python
# функция для категоризации по названиям каналов
def name_channel (row):
    if 'info' in row:
        return 'info'
    elif 'teamwork'in row:
        return 'teamwork'
    elif 'exerciser'in row:
        return 'exerciser'
    elif 'projects'in row:
        return 'projects'
    else: return 'other'
            
df['name_channel'] = df['channel'].apply(name_channel)
```

## Исследовательский анализ

### Общая активность

```python
df['client_msg_id'].count()
```

```python
# группируем по id сообщениям, и считаем количество ответов и реакций для каждого сообщения
reaction_msg = (
    df.groupby('client_msg_id')
    .agg({'latest_reply': 'count', 'cnt_reactions': 'sum'})
    .sort_values(by='cnt_reactions',ascending = False)
)
reaction_msg
```

Всего **26322** сообщения. Некоторые из сообщений не имеют id и отнесены к *unknown*

Посмотрим как часто отвечают на сообщения или ставят реакции

**Ответы**

```python
reaction_msg['latest_reply'].value_counts()
```

```python
round(reaction_msg.query('latest_reply != 0')['latest_reply'].count() / reaction_msg.shape[0] * 100, 1)
```

7.4% сообщений имеют ответы, и как правило только один ответ.

**Реакции**

```python
round(reaction_msg.query('cnt_reactions != 0')['cnt_reactions'].count() / reaction_msg.shape[0] * 100, 1)
```

22.6% сообщений имеют реакции

Посмотрим активность в тредах

```python
# группируем по времени создания треда и считаем количество уникальных сообщений, ответов и реакций
thread = (
    df.groupby('thread_ts')
    .agg({'client_msg_id' : 'nunique', 'latest_reply': 'count', 'cnt_reactions': 'sum'})
    .sort_values(by='client_msg_id',ascending = False)
)
thread
```

Всего 1540 тредов. Все треды имеют сообщения.

```python
thread['client_msg_id'].hist(bins=50)
plt.title('Распределение количества сообщений в тредах',fontsize=13)
plt.show()
```

Большая часть тредов имеют от 1 до 10 сообщений. Хотя встречаются и такие, которые имеют больше 200 сообщений.

```python
thread['latest_reply'].hist(bins=30)
plt.title('Распределение количества ответов в тредах',fontsize=13)
plt.show()
```

Посмотрим соотношение ответов и реакций с количеством сообщений. А так же постичаем вовлеченность - *er*.

```python
df_cochort = (df.groupby('cochort')
              .agg({'client_msg_id': 'count', 'latest_reply': 'count', 'cnt_reactions': 'sum', 'user': 'nunique'})
              .rename(columns = {'client_msg_id':'cnt_msg', 'latest_reply':'cnt_reply'})
             )
df_cochort['ratio_reply'] = round(df_cochort['cnt_reply'] / df_cochort['cnt_msg'], 2)
df_cochort['ratio_reactions'] = round(df_cochort['cnt_reactions'] / df_cochort['cnt_msg'], 2)
df_cochort['er'] = round((( df_cochort['cnt_reactions'] + df_cochort['cnt_reply']) / df_cochort['cnt_msg']) / df_cochort['user']* 100, 2)
df_cochort.sort_values(by='ratio_reactions', ascending=False)
```

```python
ax = df_cochort[['ratio_reactions','ratio_reply']].plot(kind='bar', figsize=(15,5))
plt.title('Распределение отношения реакций и ответов к сообщениям по когортам', fontsize=13)
ax.set_xlabel('Когорта')
ax.set_ylabel('Соотношение')
plt.legend(['отношение реакций', 'отношение ответов'])
plt.show()
```

В соотношении количества реакций к количеству сообщений лидеры поменялись.  
Однако большое количество реакций у маленьких когорт с малым количеством сообщений приводит к тому, что количество реакций превышает количество сообщений.  

Сделаем срез с соотношением меньше 1.

```python
ax = (df_cochort.query('ratio_reactions < 1')[['ratio_reactions','ratio_reply']]
      .plot(kind='bar', figsize=(15,5),width = 0.8))
plt.title('Распределение отношения реакций и ответов к сообщениям по когортам', fontsize=13)
ax.set_xlabel('Когорта')
ax.set_ylabel('Соотношение')
plt.legend(['отношение реакций','отношение ответов'])
plt.show()
```

Соотношение реакций колеблется от когорты к когорте от 0 до почти 1. 
Ответов значительно меньше, чем реакций. И не превышает 20% у всех когорт.
Лидерами стали de-22, da-plus-19 и de-5.  

При соотношении реакций к количеству сообщений большая активность все-равно остается у аналитиков и инженеров.
