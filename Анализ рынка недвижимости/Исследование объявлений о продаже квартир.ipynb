## Предобработка данных

### Стиль заголовков

Приведем название столбца `cityCenters_nearest` к единому регистру

```python
data = data.rename(columns={'cityCenters_nearest' : 'city_centers_nearest'})
```

### Поиск  и удаление дубликатов

```python
data['locality_name'].sort_values().unique()
```

Есть неявные дубликаты. Такие как "городской поселок Большая Ижора" и "поселок городского типа Большая Ижора", или "поселок городского типа Рахья" и "садовое товарищество Рахья".

```python
data['locality_name'] = data['locality_name'].replace('ё', 'е', regex=True)
data['locality_name'].sort_values().unique()
```

Удалим все обозначения населенных пунктов, оставив только названия.

```python
duplicates= ['городской поселок ', 'коттеджный поселок ', 'поселок городского типа ', 'поселок при железнодорожной станции ','поселок станции ','садовое товарищество ','поселок ','имени ', 'село ','деревня ', 'садоводческое некоммерческое товарищество ']
data['locality_name'] = data['locality_name'].replace(duplicates, '', regex=True)

data['locality_name'].sort_values().unique()
```

Проверим явные дубликаты

```python
data.duplicated().sum()
```

**Вывод:**  
Заменили неявные дубликаты - некоторые типы населенных пунктов были написаны по-разному.  
Одинаковых объявлений нет.

### Поиск и удаление пропусков

```python
data.isna().sum()
```

Выявили пропуски в 14 столбцах, как в категориальных так и в количественных значениях. Есть пропуски в данных, которые указывают владельцы - их количество значительно меньше, чем картографических.  
Пройдемся по каждому столбцу отдельно.

**Жилая площадь в квадратных метрах (м²) - `living_area`**

```python
data[data['living_area'].isna()]
```

Пропуски в значениях жилой площади не зависят от значений других столбцов. Скорее всего они случайны.

Посмотрим минимальные и максимальные значения жилой площади в зависимости от общей площади. Чтобы понять можем ли мы заменить пропуски медианным значением.

```python
data.pivot_table(index='total_area',values='living_area', aggfunc= ['min','max', 'median']).head(25)
```

Так же посмотрим корреляцию

```python
data['living_area'].corr(data['total_area'])
```

Мы видим, что корреляция между общей и жилой площадью положительная и высокая (0.93), а разброс от минимального до максимального значения не значительный.
Заменим пропуски медианным значение в зависимости от общей площади объекта.

```python
data['living_area'] = data['living_area'].fillna(data.groupby('total_area')['living_area'].transform("median"))
```

```python
data['living_area'].isna().sum()
```

Из 1903 пропусков не заполненными остались 143, так как в данных объявлениях уникальные значения общей площади. Пока оставим пропуски. Посмотрим, возможно при дальнейшей обработке данные строки будут исключены.

**Площадь кухни в квадратных метрах (м²) - `kitchen_area`**

```python
data[data['kitchen_area'].isna()]
```

Посмотрим минимальные и максимальные значения кухонной площади в зависимости от общей площади. Чтобы понять можем ли мы заменить пропуски медианным значением.

```python
data.pivot_table(index='total_area',values='kitchen_area', aggfunc= ['min','max', 'median']).head(25)
```

Корреляция

```python
data['kitchen_area'].corr(data['total_area'])
```

Корреляция меньше, чем у жилой площади с общей, и составляет 0.6. При этом видим, что разброс значений так же не велик.  
Заменим медианным значением в зависимости от общей площади.

```python
data['kitchen_area'] = data['kitchen_area'].fillna(data.groupby('total_area')['kitchen_area'].transform("median"))
data['kitchen_area'].isna().sum()
```

Из 2278 пропусков не заполненными остались 205, так как в данных объявлениях уникальные значения общей площади. Пока оставим пропуски. Посмотрим, возможно при дальнейшей обработке данные строки будут исключены.

**Высота потолков - `ceiling_height`.**

```python
data[data['ceiling_height'].isna()]
```

Мы видим, что совместно с пропусками по высоте потолков, часто встречаются пропуски в значениях других столбцов. Либо владелцы невнимательно заполняли описание, либо в сборе данных есть ошибка.

Посморим все значения высоты потолков

```python
data['ceiling_height'].sort_values().unique()
```

Стандартная высота потолков от 2.5 до 3.7 метров; в двухуровневых квартирах - 6.2 метра. Мы наблюдаем аномальные значения, такие как 26 или 100 метров. Скорее всего владелец ошибся внося параметры.  
Чтобы заменить пропуски медианным значением, сначала исправим явные несоответствия.

```python
data.loc[data['ceiling_height'] > 20.0, 'ceiling_height'] = data.loc[data['ceiling_height'] > 20.0, 'ceiling_height'] / 10
data['ceiling_height'].sort_values().unique()
```

Заменим пропуски медианным значением высоты потолков в диапазоне от 2.4 до 6.2 метров.

```python
data['ceiling_height'] =  data['ceiling_height'].fillna(data.query('2.4 < ceiling_height < 6.2')['ceiling_height'].median())
data['ceiling_height'].isna().sum()
```

**Апартаменты - `is_apartment`**

Данный столбец содержит булев тип, предполагающий ответы да или нет. Относится ли помещение к апартаментам или нет. Предположим, что если владелец не указал принадлежность к апартаментам, то скорее всего данное помещение не относится к этому типу. Поэтому заменим все пропуски на `False`.

```python
data['is_apartment'] = data['is_apartment'].fillna(False)
data['is_apartment'].isna().sum()
```

**Наличие балконов - `balcony`**

Данный столбец содержит количество балконов. Предположим, что если владелец не указал число балконов, то в квартире их нет. Поэтому заменим все пропуски нулевым значением.

```python
data['balcony'] = data['balcony'].fillna(0)
data['balcony'].isna().sum()
```

**Всего этажей в доме - `floors_total`**

Количество пропусков мало. Поэтому удалим строки, так как мы не можем знать на какое значение заменить.

```python
data = data.dropna(subset=['floors_total'])
```

**Название населённого пункта - `locality_name`**

Заменим пропуски на `unknown`

```python
data['locality_name'] = data['locality_name'].fillna('unknown')
data['locality_name'].isna().sum()
```

Пропуски значений в столбцах **`airports_nearest`, `city_centers_nearest`, `parks_around3000`, `parks_nearest`, `ponds_around3000`, `ponds_nearest`, `days_exposition`** не заполняются владелцем квартиры. Наличие большого числа пропусков и схожее количество в соответсвующих категориях, скорее всего говорит о том, что либо модель получения значений не корректно работает,либо данные были потеряны или повреждены при выгрузке.  
Заменить пропуски этих значений мы не можем. А при удалении потеряеться больше 20% данных. Поэтому оставим пропуски без изменений.

### Изменение типов данных

```python
data.dtypes
```

В столбце `first_day_exposition` даты представлены в формате `object`. Для того чтобы можно было делать вычисления и преобразования, переведем их в формат `datetime64`.

```python
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')
data['first_day_exposition'].head()
```

В столбце `last_price` цена представлена в формате `float64`. Переведем в формат `int64`.

```python
data['last_price'] = data['last_price'].astype('int64')
```

В столбце `floors_total` количество этажей представлены в формате `float64`. Переведем их в формат `int64`.

```python
data['floors_total'] = data['floors_total'].astype('int64')
```

В столбце `balcony` количество балконов представлены в формате `float64`. Переведем их в формат `int64`.

```python
data['balcony'] = data['balcony'].astype('int64')
```

```python
data.dtypes
```

## Посчитайте и добавьте в таблицу новые столбцы

**Цена одного квадратного метра - `price_square_meter`**

```python
data_sort['price_square_meter'] = data_sort['last_price'] / data_sort['total_area']
data_sort.head()
```

**День недели публикации объявления - `week_day`**

```python
data_sort['week_day'] = data_sort['first_day_exposition'].dt.weekday
data_sort.head()
```

**Mесяц публикации объявления -`month_publication`**

```python
data_sort['month_publication'] = data_sort['first_day_exposition'].dt.month
data_sort.head()
```

**Год публикации объявления - `year_publication`**

```python
data_sort['year_publication'] = data_sort['first_day_exposition'].dt.year
data_sort.head()
```

**Тип этажа квартиры - `type_floor`**

Напишем функцию, для классификации.

```python
try:
    def type_floor(row):
        floor = row['floor']
        floors_total = row['floors_total']
        if floor == 1:
            return 'первый'
        if floor == floors_total:
            return 'последний'
        return 'другой'

    data_sort['type_floor'] = data_sort.apply(type_floor, axis=1)
except:
    'Проверь значения'
    
data_sort.head()
```

```python
data_sort['city_center_kilometers'] = round(data_sort['city_centers_nearest'] / 1000)
data_sort.head()
```
