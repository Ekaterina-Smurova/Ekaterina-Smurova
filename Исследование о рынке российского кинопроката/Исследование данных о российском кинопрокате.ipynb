```python
import pandas as pd
import matplotlib.pyplot as plt
data_1 = pd.read_csv('/datasets/mkrf_movies.csv')
data_2 = pd.read_csv('/datasets/mkrf_shows.csv')
```

Объединим обе таблицы по номеру прокатного удостоверения.

Проверим, что по ключу объединения нет пропусков и дубликатов.

```python
data_1['puNumber'].isna().sum()
```

```python
data_2['puNumber'].isna().sum()
```

Выявили 2 дубликата по номеру прокатного удостоверения в первом датафрейме.

```python
data_copy['puNumber'].sort_values().unique()
```

```python
data_copy = data_copy[data_copy['puNumber'] != 'нет']
```

```python
data_copy[data_copy['puNumber'] == 'нет']
```

```python
data_copy['puNumber'] = data_copy['puNumber'].astype('int')
```

## Шаг 2. Предобработка данных

### Стиль заголовков

Приведем название столбца `puNumber` к единому стилю

```python
data = data.rename(columns={'puNumber' : 'pu_number'})
```

Перед дальнейшими дейсвиями сделаем копию данных, чтобы потом оценить сколько мы удалили.

```python
df = data.copy()
```

### Изучите пропуски в датафрейме

- Объясните, почему заполнили пропуски определённым образом или почему не стали этого делать.

```python
df.isna().sum()
```

Выявлены пропуски в 11 столбцах,как в категориальных так и в количественных значениях.  
Пропуски в столбцах `refundable_support`, `nonrefundable_support`, `budget`, `financing_source` идентичны. Что логично, так как значения этих столбцов относятся к фильмам получавшим гос.поддержку.  
Пройдемся по каждому столбцу отдельно.

#### **Студия-производитель - `film_studio`**

```python
df[df['film_studio'].isna()]
```

Пропуски в значениях студии-производителя не зависят от значений других столбцов. Скорее всего они случайны.  
Так как в дальнейшем исследовании эти значения не используются, заменим пропуски на `unknown`.

```python
df['film_studio'] = df['film_studio'].fillna('unknown')
```

#### **Страна-производитель - `production_country`**

```python
df[df['production_country'].isna()]
```

Мы видим пропуски стрыны в названиях Ералаш и Мульт личности. Значит можем заменить пропуски на Россию.

Посмотрим распределение гос.поддержки и кассовых сборов.

```python
#Выявим характерные значения кассовых сборов
df.query('box_office > 0')['box_office'].describe()
```

```python
#Напишем функцию для категоризации кассовых сборов
q1 = df['box_office'].quantile(0.25)
q2 = df['box_office'].quantile(0.5)
q3 = df['box_office'].quantile(0.75)


def category(value):
    try:
        if value <= q1:
            return '1-низкий'
        if value <= q2:
            return '2-ниже среднего'
        if value <= q3:
            return '3-средний'
        if value <= 1500000000:
            return '4-выше среднего'
        return '5-высокий'
    except:
        'Что-то пошло не так'
        
df['category_box_office'] = df.query('box_office > 0')['box_office'].apply(category)
```

Окупиться смогли только фильмы со сборами выше 1.5млрд.  

Чем больше кассовые сборы, тем выше окупаемость. Но нельзя точно сказать, что это единственный показатель, который влияет на окупаемость.
Возможно данные фильмы дольше находились в прокате или выходили в праздничные дни.

Частично окупились фильмы 2018г.
В остальные года фильмы не окупались.  

Фильмы с большей окупаемостью выходили в 2017г., 2018г. и 2019г.  
Прокат стал окупатся только в 2014г. и окупаемость постепенно увеличивалась.  
Возможно киноиндустрия развивалась по годам вцелом: от улучшения качества съемок и спецэффектов, до более комфортных кинозалов и рекламы.

Больше всего выделяются средства на фильмы в жанре "история", хотя общий бюджет при этом превышает кассовые сборы.
Меньше всего выделют фильмам в жанре "мелодрамма".
Окупаются только фильмы в жанре "спорт".  

Самые окупаемые фильмы были в жанрах: "комедия", "спорт" и "драма".  

Большая окупаемость в категориях "6+" и "12+".
Что вполне логично, на такие фильмы обычно ходят всей семьей.  

**Вывод**  
Исходные данны были слишком загрязнены, что влияло на анализ.  

С 2016г. увеличилось количество фильмов вышедших в прокат, и сильно возросли кассовые сборы. 

Так же постепенно с 2014г. фильмы стали выходить на окупаемость.

Жанры "спорт" и "комедия" пожалуй самые популярные у нас в стране.  

На окупаемость влияют:  
* доля возвратных средств от общего бюджета. Чем меньше возвратных средств, тем выше окупаемость.  
* жанр.  Наиболее окупаемые жанры: "комедия", "спорт" и "драма".
* возрастная категория. Большая окупаемость в категориях "6+" и "12+".  

На развитие кинематографа выделяются огромные бюджеты. Однако фильмы с гос.поддержкой набирают средние рейтинги, и часто не окупаются.
