**Цель проекта:** снизить отток клиентов, подобрав для них оптимальные тарифы  

**Заказчик:**  отдел по работе с клиентами провайдера виртуальной телефонии «Нупозвони»  

**Задачи:** 
 - провести исследовательский анализ данных;
 - получить расходы клиентов на каждом из тарифов;
 - выделить клиентов, для которых оптимальный тариф не совпадает с текущим (переплачивают на текущих тарифах; сэкономят, если перейдут на более дорогой тариф);
 - показать изменение в выручке для заказчика (общей, в месяц, на клиента) при смене тарифов;
 - проанализировать полученные результаты, по возможности дать общее понимание ситуации (каким клиентам более выгоден тот или иной тариф);
 - проверить статистические гипотезы:
      * Уровень оттока клиентов при использовании оптимального и неоптимального тарифного плана не отличается
      * Выручка провайдера не изменится после переключения всех пользователей на оптимальные для них тарифы
      * Выручка провайдера не изменится после переключения переплачивающих пользователей на более "дорогие" для них тарифы



**Правила формирования тарифов**

Расходы клиента состоят из:

* ежемесячной платы +
* платы за перерасход по внутренним исходящим звонкам, с учетом времени ожидания +
* платы за внешние исходящие звонки, с учетом времени ожидания +
* платы за оператора

Счета выставляются за прошедший календарный месяц, поэтому датой списания можно считать 1-е число каждого месяца. Остаток трафика сгорает на начало месяца.  

Тарификация происходит на основе кумулятивных показателей за период расчета (календарный месяц). Общая продолжительность звонков (в минутах) за день по каждому оператору округляется в большую сторону. 

|Тарифный план|Ежемесячная плата|Входящие|Исходящие звонки внутри сети, включённые в тариф|Цена сверх лимита|Исходящие звонки, кроме внутренних|Дополнительная ежемесячная плата за каждого оператора|
|----|----|----|----|----|----|----|
|Тариф А|4000 ₽|0|2000 минут|0.1 ₽/мин.|0.3 ₽/мин.|50 ₽|
|Тариф B|2000 ₽|0|500 минут|0.15 ₽/мин.|0.5 ₽/мин.|150 ₽|
|Тариф С|1000 ₽|0|нет|0.3 ₽/мин.|0.7 ₽/мин.|300 ₽|

**Описание данных:**  

Датасеты описывают использование услуг «Нупозвони» — провайдера
виртуальный телефонии.  
Его клиенты — колл-центры, которые:
* распределяют входящие вызовы на операторов,
* совершают исходящие вызовы силами операторов.  

Также операторы могут делать внутренние вызовы — вызовы друг между другом
внутри сети виртуальной телефонии.  

В нашем распоряжении два датасета.  

Колонки в **telecom_clients**:  

`user_id` - идентификатор клиента  
`tariff_plan` - тарифный план клиента  
`date_start` - дата регистрации клиента в сервисе  

Колонки в **telecom_dataset**:  

`user_id` - идентификатор клиента  
`date` - дата статистики  
`direction` - направление вызовов: `out` - исходящий вызов, `in` - входящий  
`internal` - маркер внутренних и внешних вызовов  
`operator_id` - идентификатор оператора  
`is_missed_call` - маркер пропущенных вызовов  
`calls_count` - количество звонков  
`call_duration` - длительность звонка без учёта времени ожидания  
`total_call_duration` - длительность звонка с учётом времени ожидания

### Расходы клиентов на каждом из тарифов

**Создадим таблицу с тарифами**

```python
# создаем таблицу с тарифами
tariffs = pd.DataFrame(
    data={'tariff_plan' : ['A', 'B', 'C'], 
          'monthly' : [4000, 2000, 1000],
          'minutes_included' : [2000, 500, 0],
          'internal_out' : [0.1 , 0.15 , 0.3], 
          'external_out' : [0.3, 0.5, 0.7], 
          'operator' : [50, 150, 300]}, 
)

tariffs
```

Тарификация происходит на основе кумулятивных показателей за период расчета (календарный месяц).  
Общая продолжительность звонков (в минутах, с учетом времени ожидания) за день по каждому оператору округляется в большую сторону.  

Произведем расчет количества операторов для каждого клиента с учетом месяца получения метрики и добавим в таблицу.

```python
# Вычисление количества операторов для каждого клиента
count_operator = actual_df.groupby(['user_id', 'month_metric'],as_index=False).agg({'operator_id': 'nunique'})\
                        .rename(columns={'operator_id': 'count_operator'})
# Добавление столбца с количеством операторов для каждого клиента в таблицу
actual_df = actual_df.merge(count_operator, on=['user_id', 'month_metric'], how='left')
```

Формируем итоговую таблицу по месяцам для каждого клиента.  


```python
df_month = (actual_df.query('direction !="входящий"')
            .pivot_table(index=['user_id', 'tariff_plan',  'month_metric', 'internal', 'count_operator'],
                                   values='total_call_duration',aggfunc='sum').reset_index()
           )
df_month 
```

Рассчитаем гипотетические затраты на каждом из тарифов

```python
# Добавляем столбцы с тарифами
for i, tariff in enumerate(tariffs['tariff_plan']):
    df_month[tariff] = 0
df_month

#Для каждого клиента считаем гипотетические затраты по каждому вызову
for row in  df_month:
    df_month.loc[df_month['internal'] == True & (df_month['total_call_duration'] > tariffs.loc[0,'minutes_included']), 'A'] = \
    (df_month['total_call_duration'] - tariffs.loc[0,'minutes_included']) * tariffs.loc[0,'internal_out']
    
    df_month.loc[df_month['internal'] == True & (df_month['total_call_duration'] > tariffs.loc[1,'minutes_included']), 'B'] = \
    (df_month['total_call_duration'] - tariffs.loc[1,'minutes_included']) * tariffs.loc[1,'internal_out']
    
    df_month.loc[df_month['internal'] == True & (df_month['total_call_duration'] > tariffs.loc[2,'minutes_included']), 'C'] = \
    (df_month['total_call_duration'] - tariffs.loc[2,'minutes_included']) * tariffs.loc[2,'internal_out']
    
    df_month.loc[df_month['internal'] == False, 'A'] = df_month['total_call_duration'] * tariffs.loc[0,'external_out']
    df_month.loc[df_month['internal'] == False, 'B'] = df_month['total_call_duration'] * tariffs.loc[1,'external_out']
    df_month.loc[df_month['internal'] == False, 'C'] = df_month['total_call_duration'] * tariffs.loc[2,'external_out']
    df_month[tariff] = df_month[tariff].fillna(0)
df_month
```

```python
#Для каждого клиента объединяем затраты по всем звонкам
df_month_price = df_month.pivot_table(index=['user_id','tariff_plan','month_metric','count_operator'], values=['A','B','C'], aggfunc='sum').reset_index()
df_month_price
```

```python
# Прибавляем стоимость месяца и стоимость за операторов
df_month_price['A'] = df_month_price['A'].apply(lambda x: x + tariffs.loc[0,'monthly'] + (df_month_price['count_operator'] * tariffs.loc[0,'operator']))
df_month_price['B'] = df_month_price['B'].apply(lambda x: x + tariffs.loc[1,'monthly'] + (df_month_price['count_operator'] * tariffs.loc[1,'operator']))
df_month_price['C'] = df_month_price['C'].apply(lambda x: x + tariffs.loc[2,'monthly'] + (df_month_price['count_operator'] * tariffs.loc[2,'operator']))
df_month_price.drop(['count_operator'], axis= 1 , inplace= True )
df_month_price
```

Добавим столбец с фактическими затратами.

```python
# Создадим функцию, которая возвращает значение соответствующее тарифу
def actual(row):
    if row['tariff_plan'] == 'A':
        return row['A']
    if row['tariff_plan'] == 'B':
        return row['B']
    if row['tariff_plan'] == 'C':    
        return row['C']

df_month_price['actual_costs'] = df_month_price.apply(actual, axis = 1)
df_month_price
```

Добавим столбец с подходящим тарифом

```python
# Определение выгодного тарифа
df_month_price['best_tariff'] = df_month_price[['A', 'B', 'C']].idxmin(axis=1)
df_month_price
```

Добавим столбец с маркером соответствия выбранного тарифа оптимальному.

```python
def coincide(row):
    if row['tariff_plan'] == row['best_tariff']:
        return "совпадает"
    else:
        return "не совпадает"

df_month_price['coincide_tariff'] = df_month_price.apply(coincide, axis = 1)
df_month_price
```

### Проверка гипотез

#### Уровень оттока клиентов при использовании оптимального и неоптимального тарифного плана не отличается.

H0 гипотеза: доля клиентов ушедших в отток при использовании оптимального тарифа не отличается от доли клиентов ушедших в отток при использовании неоптимального тарифа.  
H1 гипотеза: - доля клиентов ушедших в отток при использовании оптимального тарифа отличается от доли клиентов ушедших в отток при использовании неоптимального тарифа.  

Сформируем таблицу с общим числом клиентов, активных и неактивных по совпадению тарифа.

```python
churn = distribution_by_tariffs.pivot_table(index='coincide_tariff', columns ='active', values ='user_id', aggfunc= 'nunique')
churn.columns=['active','inactive']
churn['count_user'] = churn['active'] + churn['inactive']
churn['%churn'] = churn['inactive'] / churn['count_user'] * 100 
churn
```

Проверку гипотезы выполним методом двустороннего z-теста.

```python
alpha = 0.05
count = np.array([churn['inactive'][0], churn['inactive'][1]])
nobs = np.array([churn['count_user'][0], churn['count_user'][1]])
stat, p_value = proportions_ztest(count, nobs,  alternative='two-sided')
if p_value < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Нет статистической разницы между оттоками клиентов при использовании оптимального и неоптимального тарифа.

#### Выручка провайдера  не изменится после переключения всех пользователей на оптимальные для них тарифы

Расчет выполним для последнего месяца получения выручки, предположив, что клиенты воспроизведут похожие метрики в следующем месяце:

H0 гипотеза: статистически значимых различий в выручке нет после смены тарифа  
H1 гипотеза: статистически значимые различий в выручке есть после смены тарифа

```python
sample_1 = distribution_by_tariffs.query('month_metric ==11')['actual_costs']
sample_2 = distribution_by_tariffs.query('month_metric ==11')['probable_costs']
```

Проведем тест Шапиро-Уилка для проверки выборок на нормальное распределение.

```python
alpha = 0.05  # критический уровень статистической значимости

results = st.shapiro(sample_1)
p_value = results[
    1
]  # второе значение в массиве результатов (с индексом 1) - p-value

print('p-значение: ', p_value)

if p_value < alpha:
    print('Отвергаем нулевую гипотезу: распределение не нормально')
else:
    print('Не получилось отвергнуть нулевую гипотезу, всё нормально') 
```

```python
alpha = 0.05  # критический уровень статистической значимости

results = st.shapiro(sample_2)
p_value = results[
    1
]  # второе значение в массиве результатов (с индексом 1) - p-value

print('p-значение: ', p_value)

if p_value < alpha:
    print('Отвергаем нулевую гипотезу: распределение не нормально')
else:
    print('Не получилось отвергнуть нулевую гипотезу, всё нормально') 
```

Так как распределение выборок не нормально, тогда необходимо применить критерий Уилкоксона-Манна-Уитни для выявления статистически значимого различия в выборках.

```python
# критический уровень статистической значимости
alpha = 0.05

results = st.mannwhitneyu(sample_1, sample_2)

print("alpha =",0.05)
print("p-value: {0:.3f}".format(results[1]))

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print(
        'Не получилось отвергнуть нулевую гипотезу')
    
mean_val = (sample_2.mean()/ (sample_1.mean()) - 1) * 100
print(f'Относительные изменения после смены тарифа: {mean_val:.4} %')
```

Различия в выручке статистически значимые. При смене всех пользователей на оптимальный тариф, выручка снизиться на 20.43%

#### Выручка провайдера не изменится после переключения переплачивающих пользователей на более дорогие для них тарифы

H0 гипотеза: статистически значимых различий в выручке нет после смены тарифа
H1 гипотеза: статистически значимые различий в выручке есть после смены тарифа

```python
sample_1 = distribution_by_tariffs.query('month_metric ==11')['actual_costs']
sample_2 = distribution_by_tariffs.query('month_metric ==11')['expensive_cost']
```

```python
# критический уровень статистической значимости
alpha = 0.05

results = st.mannwhitneyu(sample_1, sample_2)

print("alpha =",0.05)
print("p-value: {0:.3f}".format(results[1]))

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print(
        'Не получилось отвергнуть нулевую гипотезу')
    
mean_val = (sample_2.mean()/ (sample_1.mean()) - 1) * 100
print(f'Относительные изменения после смены тарифа: {mean_val:.4} %')
```
